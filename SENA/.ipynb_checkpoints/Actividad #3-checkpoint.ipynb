>#### Agrupar los datos es una parte importante para el proceso de manipulacion de datos debido a que permite una mayor claridad en el analisis.
>#### En este caso utilizaremos el rango de edad ya que es una de las variables objeto de analisis, por lo tanto procedemos a definir los rangos de edades

```python
rangos = [17,28,38,48,58,68]
```

>#### Establecemos el nombre para cada rango

```python
nombreRango = ['A','B','C','D','E']
```

>#### Creamos una nueva columna llamada _**Rango_Edad**_ donde agruparemos los datos segun el rango de edad previamente definido

```python
df['Rango_Edad'] = pd.cut(df['edad'], rangos, labels=nombreRango)
```

>#### Otra medida que se puede aplicar para el análisis de datos es el uso de diferentes tipos de _gráficos_, los cuales permiten una visualización más clara. Con el uso de los gráficos, se puede realizar una distribución mucho más _efectiva de los datos, lo que _facilita su interpretación_

>#### Para determinar el total de valores porcentualmente pagados al seguro por personas no fumadoras y fumadoras, se usa el siguiente comando, el cual genera un gráfico de total porcentual. 

```python
fig, ax = plt.subplots()
fumador = ['no', 'yes']
cantidad = df['fumador'].groupby(df['fumador']).count()
bar_labels = ['NO', 'YES']
bar_colors = ['tab:red', 'tab:green']
ax.bar(fumador, cantidad, label=bar_labels, color=bar_colors)
ax.set_ylabel('# de Personas')
ax.set_title('Cantidad de Fumadores')
plt.show()
```

```python
total_fumador = df['fumador'].groupby(df['fumador']).count()
labels = ['NO', 'YES']
plt.pie(total_fumador, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
total_rango_edad = df['Rango_Edad'].groupby(df['Rango_Edad'], observed=True).count()
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(total_rango_edad, labels=labels, autopct='%.1f%%')
plt.show()
```
