```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medical_examination.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
fumador = ['no', 'yes']
cantidad = df['fumador'].groupby(df['fumador']).count()
bar_labels = ['NO', 'YES']
bar_colors = ['tab:red', 'tab:green']
ax.bar(fumador, cantidad, label=bar_labels, color=bar_colors)
ax.set_ylabel('# de Personas')
ax.set_title('Cantidad de Fumadores')
plt.show()
```

```python
#Grafico de correlacion
correlacion = df.corr(numeric_only=True)
sns.heatmap(correlacion, xticklabels=correlacion.columns, yticklabels=correlacion.columns, annot=True)
```
